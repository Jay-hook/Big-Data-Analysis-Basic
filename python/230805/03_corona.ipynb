```python
# Unnamed: 0 컬럼을 삭제
# drop(지워질 컬럼이나 인덱스, axis=0,1{'rows', 'columns'})
# inplace 매개변수의 기본값은 False
# inplace의 값을 True로 변경하면 기준이 되는 데이터프레임을 변경
corona.drop('Unnamed: 0', axis='columns', inplace=True)
# corona = corona.drop("Unnamed: 0", axis=1)
```

```python
# 인덱스가 0인 데이터를 삭제 
# corona = corona.drop(0, axis=0)
corona.drop(0, axis='rows', inplace=True)
```

```python
# corona  데이터프레임에서 컬럼의 이름을 변경
corona.columns = ['등록일시', '누적사망자', '누적확진자', '게시글번호', 
                  '기준일', '기준시간', '수정일시', '누적의심자', 
                  '누적확진율']
```

```python
corona.head(3)
```

```python
# 데이터 프레임의 정렬을 변경
# sort_values()
# 매개변수 ascending : True(기본값)(오름차순), False(내림차순)
# 매개변수 inplace : False(기본값)(데이터의 변화 없음), True(기준이 되는 데이터프레임 변경)
corona.sort_values('등록일시', inplace=True)
# corona = corona.sort_values('등록일시')
```

```python
corona.head(3)
```

```python
# 인덱스를 초기화
# reset_index()
# 매개변수 drop : False(기본값)(기존의 인덱스를 컬럼으로 이동)
#               True(기존의 인덱스를 삭제)
# 매개변수 inplace 
corona.reset_index(drop=True, inplace=True)
# corona = corona.reset_index().drop('index', axis=1)
```

```python
corona.head(5)
```

```python
# 행을 이동하는 함수
# shift(n) : n만큼 행을 이동 n의 기본값은 1
corona['일일확진자'] =  \
    corona['누적확진자'] - corona['누적확진자'].shift()
```

```python
corona.head(3)
```

```python
# 데이터 간의 차이를 출력하는 함수
# diff(n)
# n의 기본값은 1 -> n만큼 행이 이동한 값과의 차이 
corona['일일사망자'] = corona['누적사망자'].diff()
```

```python
corona.info()
```

```python
# 결측치를 확인하는 함수
# isna()
corona.isna().sum()
```

```python
# 결측치를 채우는 함수
# fillna()
# 인자값으로 결측치를 대체
# method 매개변수를 이용하여 대체
# ffill : 전 행의 데이터로 결측치를 대체
# bfill : 후 행의 데이터로 결측치를 대체

corona[['일일확진자', '일일사망자']] = \
    corona[['일일확진자', '일일사망자']].fillna(0)
```

```python
corona['누적의심자'].fillna(method='bfill')
```

- iloc[]
    - [행의 위치, 열의 위치] : 해당하는 위치의 데이터를 출력
    - 행의 위치만 있는 경우 [행의 위치]
    - 열의 위치만 있는 경우 [:, 열의 위치]

```python
# 인덱스의 값을 초기화 (기존의 인덱스는 삭제)
corona.reset_index(drop=True, inplace=True)
# corona = corona.reset_index(drop=True)
```
