```python
import pandas as pd
```

### 연습문제 
1. csv 폴더에 있는 sales records.csv 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel' 컬럼을 삭제 
3. 컬럼의 이름 변경 -> ['대륙', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 대륙과 국가를 기준으로 오름차순 정렬
5. 기존의 인덱스를 삭제하고 인덱스를 초기화 
6. 새로운 파생 변수 생성
    - 총단가 -> 판매단위 * 단가
    - 총원가 -> 판매단위 * 원가
    - 총이윤 -> 총단가 - 총원가

```python
sales = pd.read_csv("../../csv/Sales Records.csv")
sales.head(1)
```

```python
sales.info()
```

```python
# 컬럼의 삭제 
sales.drop(['Order ID', 'Total Cost', 'Total Revenue', 
            'Total Profit', 'Sales Channel'], axis=1, inplace=True)

# sales = sales.drop(['Order ID', 'Total Cost', 'Total Revenue', 
#             'Total Profit', 'Sales Channel'], axis=1)
```

```python
# 컬럼의 이름을 변경 
sales.columns = ['대륙', '국가', '상품종류', '우선순위', '주문일자', 
                 '발송일자', '판매단위', '단가', '원가']
```

```python
# 대륙과 국가를 기준으로 오름차순 정렬
sales.sort_values(['대륙', '국가'], inplace=True)
# sales = sales.sort_values(['대륙', '국가'])
```

```python
# 기존의 인덱스를 제거하고 인덱스를 초기화
sales.reset_index(drop=True, inplace=True)
# sales = sales.reset_index(drop=True)
```

```python
# 새로운 파생변수 3개를 생성

# 총단가 -> 단가 * 판매단위
sales['총단가'] = sales['단가'] * sales['판매단위']
# 총원가 -> 원가 * 판매단위
sales['총원가'] = sales['원가'] * sales['판매단위']
# 총이윤 -> 총단가 - 총원가
sales['총이윤'] = sales['총단가'] - sales['총원가']
```

```python
sales.info()
```

```python
# 빈도수를 체크하는 함수 
sales['국가'].value_counts()
```

```python
## 국가가 한국(South Korea)인 데이터만 출력 
sales.loc[sales['국가'] == 'South Korea']
```

1. 국가별 총이윤의 합계
    - 그룹화가 아닌 방법을 이용하여 국가별 총이윤 합계
    - 그룹화 연산

```python
sales.groupby('국가').sum()[['총이윤']]
```

```python
sales[['국가', '총이윤']].groupby('국가').sum()
```

```python
## 국가 컬럼의 데이터중 중복된 값은 삭제하고 국가의 리스트 생성
sales['국가'].value_counts().index
```

```python
## 국가 데이터에서 중복된 값은 삭제하고 리스트화 함수
_list = sales['국가'].unique()
```

```python
flag = sales['국가'] == 'South Korea'
df = sales.loc[flag]
df['총이윤'].sum()
```

```python
# 반복문을 이용하여 국가별로 총이윤의 합계 출력

# 비어있는 리스트 생성
values = []

for i in _list:
    # print(i)
    flag = sales['국가'] == i
    df = sales.loc[flag]
    values.append(round(df['총이윤'].sum(), 2))

values
```

```python
pd.DataFrame(values, index=_list, columns=['총이윤']).sort_index()
```
